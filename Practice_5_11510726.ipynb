```python
from sklearn import tree

attributes = iris_df[["sepal length","sepal width","petal length","petal width"]]
target = iris_df[["class"]]

clf = tree.DecisionTreeClassifier(criterion='entropy')
clf = clf.fit(attributes,target)
```

```python
clf.predict([[1,1,1,1]])
```

```python
prediction = clf.predict(attributes)
prediction_df = pd.DataFrame({"prediction": prediction})
prediction_df.head()

training_result = pd.concat([prediction_df, target], axis=1)
training_result.head()
```

```python
misclass = 0 
for i in range(len(training_result)):
    if training_result['prediction'][i]==training_result['class'][i]:
        misclass=misclass+1
print("Misclassifications rate is")
print(1-(misclass/len(training_result)))
```

```python
url_1 = 'https://archive.ics.uci.edu/ml/machine-learning-databases/haberman/haberman.data'

huberman=pd.read_csv(url_1, names=("AgeOfPatient","PatientYear","NumbersPositive","Status"))
huberman[:].head()
```

```python
huberman.describe()
```

```python
fig1 = plt.figure()
ax1 = Axes3D(fig1)

x1 = huberman["AgeOfPatient"]
y1 = huberman["PatientYear"]
z1 = huberman["NumbersPositive"]
huberman["Status"]
colors1 = {1:'red', 2:'blue'}

ax1.scatter(x1, y1, z1, zdir='z1', c = huberman["Status"].apply(lambda x1: colors1[x1]))

ax1.set_xlabel('AgeOfPatient')
ax1.set_ylabel('PatientYear')
ax1.set_zlabel('NumbersPositive')

plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
attributes = huberman[["AgeOfPatient","PatientYear","NumbersPositive"]]
target = huberman[["Status"]]

cft = tree.DecisionTreeClassifier(criterion='entropy')
X_train,X_test, y_train, y_test = train_test_split(attributes,target,test_size=0.4, random_state=0) 
cft = cft.fit(X_train,y_train)

predict = cft.predict(X_test)
accuracy = accuracy_score(y_test, predict,normalize=True)
print(accuracy)
print()
print('Misclassification rate is:')
print(1-accuracy)
```
